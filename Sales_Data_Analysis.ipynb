# 2:Which product had the highest revenue? How much revenue did it generate?

```python
newdf = df.groupby("product")[['revenue']].sum()
newdf2 = newdf.max()
result = newdf.idxmax()
# Display result
print(newdf2,"product had the highest revenue:\n",result)
```

```python
import matplotlib.pyplot as plt
# Plot the revenue of each product as a bar chart
newdf.plot(kind="bar", rot=27)
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.title("Revenue by Product")
plt.show()
```
